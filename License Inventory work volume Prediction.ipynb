```python
from statsmodels.tsa.stattools import adfuller

def test_stationary(timeseries):
    
    #Determining the rolling statistics
    mvgavg=timeseries.rolling(window=12).mean()
    mvgstd=timeseries.rolling(window=12).std()
    
    #plotting the rolling statistics
    plt.plot(timeseries, label='original')
    plt.plot(mvgavg, color='red',label='moving average')
    plt.plot(mvgstd, color='black', label='standard deviation')
    plt.title('moving average # standard deviation')
    plt.legend(loc='best')
    plt.show(block=False)
    
    
    ad_test=adfuller(timeseries['Request_Count'])
    ad_output=pd.Series(ad_test[0:4], index=['Test statistics','p-value','#lag used','no of observation used'])
    
    for key, value in ad_test[4].items():
        ad_output['critical value(%s)'%key]=value
    print(ad_output)
    
```

```python
LI_PDC_log_shift=LI_PDC_log-LI_PDC_log.shift(1)
```

```python
plt.plot(LI_PDC_log_shift)
```

```python
LI_PDC_log_shift.dropna(inplace=True)
test_stationary(LI_PDC_log_shift)
```

```python
#ACF and PACF plot:
from statsmodels.tsa.stattools import acf, pacf

lag_acf=acf(LI_PDC_log_shift, nlags=20)
lag_pacf=pacf(LI_PDC_log_shift, nlags=20, method='ols')

#plot ACF:
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(LI_PDC_log_shift)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(LI_PDC_log_shift)),linestyle='--',color='gray')
plt.title('autocorrelation function')

#plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(LI_PDC_log_shift)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(LI_PDC_log_shift)),linestyle='--',color='gray')
plt.title('partial autocorrelation function')

plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA

#AR model
model=ARIMA(LI_PDC_log, order=(2,1,0))
results_AR=model.fit(disp=-1)
plt.plot(LI_PDC_log_shift)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'%sum(results_AR.fittedvalues-LI_PDC_log_shift['Request_Count'])**2)

print('Plotting AR model')
```

```python
model=ARIMA(LI_PDC_log, order=(0,1,2))
results_ma=model.fit(disp=-1)
plt.plot(LI_PDC_log_shift)
plt.plot(results_ma.fittedvalues, color='red')
plt.title('RSS: %.4f'%sum(results_ma.fittedvalues-LI_PDC_log_shift['Request_Count'])**2)

print('Plotting MA model')
```

```python
model=ARIMA(LI_PDC_log, order=(2,1,2))
results_arima=model.fit(disp=-1)
plt.plot(LI_PDC_log_shift)
plt.plot(results_arima.fittedvalues, color='red')
plt.title('RSS: %.4f'%sum(results_arima.fittedvalues-LI_PDC_log_shift['Request_Count'])**2)

print('Plotting AR model')
```
